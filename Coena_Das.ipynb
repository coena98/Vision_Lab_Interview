### 2. You have collected data on the growth rate of cancer cells and built a statistical model that fits your acquired data well. What general strategies exist to test your model? 

There are several general strategies that can be used to test a statistical model:

Holdout method: This involves dividing the data into a training set and a testing set. The model is trained on the training set and then tested on the testing set. This allows the developer to evaluate the model's performance on unseen data.

Cross-validation: This involves dividing the data into a number of folds (e.g. 5 folds) and training the model on a different fold each time, while using the remaining folds as the test set. This allows the developer to get a better estimate of the model's performance, as it is tested on multiple different sets of data.

Bootstrapping: This involves sampling the data with replacement to create multiple bootstrapped samples. A model is trained on each bootstrapped sample and the results are combined to estimate the model's performance.

Simulation: If the data generating process is known, it is possible to simulate data using this process and use the simulated data to test the model. This allows the developer to evaluate the model's performance under different conditions.

Evaluate the model's fit: It is also important to evaluate how well the statistical model fits the data. Using measures such as the coefficient of determination (R^2) or the root mean squared error (RMSE), the developer can assess the model's fit. In case if the model is used for classification, metrics such as accuracy, precision, recall, F1 Score, CLassification Matrix and ROC curve can be used to evaluate it. 

Check for model assumptions: Finally, the developer should also check that the statistical model makes reasonable assumptions about the data. For example, if the model assumes that the errors are normally distributed, Q-Q plot or a goodness-of-fit test can be used to check this assumption.

### **Extracting pixel information for each band in a Pandas Dataset**

```python
def create_pixel_dataset(img_data, labels):
    df = pd.DataFrame()
    for i in tqdm(range(img_data.shape[2])):
        df = pd.concat([df, pd.DataFrame(img_data[:, :, i].ravel())], axis=1)
    df = pd.concat([df, pd.DataFrame(labels.ravel())], axis=1)
    df.columns = [f'band-{i}' for i in range(1, 1+img_data.shape[2])]+['class']
    return df
```

### **Performing k-means clustering of the data with k=16 classes and visualising the resulting scene classification.**

```python
# Perform k-means clustering with k=16 clusters
kmeans = KMeans(n_clusters=16).fit(X)

# Predict the cluster labels for each data point
y_pred = kmeans.predict(X)
```

```python
#scatter plot of the Kmeans Clustering
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.show()
```

```python
# Visualize the resulting scene classification
plt.figure(figsize=(8, 6))
plt.imshow(y_pred.reshape(145, 145), cmap='viridis')
plt.title('Kmeans', fontsize=14)
plt.colorbar()
plt.show()
```

### **Normalising the input spectra to unit norm and then visualizing resulting scene classification**

```python
# Normalize the features to have unit norm
X_norm = Normalizer().fit_transform(X)

kmeans_norm = KMeans(n_clusters=16).fit(X_norm)

# Predict the cluster labels for each data point
y_pred_norm = kmeans_norm.predict(X_norm)
```

```python
#scatter plot of Kmeans Clustering with unit normalization
plt.scatter(X[:, 0], X[:, 1], c=y_pred_norm, s=50, cmap='viridis')
centers = kmeans_norm.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

```python
# Visualize the resulting scene classification
plt.figure(figsize=(8, 6))
plt.imshow(y_pred_norm.reshape(145, 145),cmap='viridis')
plt.title('Kmeans Normalized', fontsize=14)
plt.colorbar()
plt.show()
```

Unit Normalization makes the clustering more robust to the presence of outliers or large variations in the scale of the data, which can lead to better separation between clusters.The result leads to more clear demarcation of classes, where classes are more homogenous. Better demarcation of classes refers to a clearer separation and more distinct boundaries between the different classes of land cover in this scene.
